```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
project_dir = os.getcwd()  
file_path = os.path.join(project_dir, "Sample_Data_for_Activity.csv")


df = pd.read_csv(file_path)
df.head()
```

```python
save_dir = "Output"
os.makedirs(save_dir, exist_ok=True)
```

```python
# 📊 1️⃣ 直方图（Histogram）：查看数据分布情况
plt.figure(figsize=(12, 8))
df.hist(bins=30, alpha=0.7, grid=False, edgecolor="black", layout=(2,2), figsize=(12,8))
plt.suptitle("Histograms of Different Distributions", fontsize=14)
plt.savefig(os.path.join(save_dir, "histograms.png"), dpi=300, bbox_inches="tight")
plt.show()
```

```python
# 📊 2️⃣ 箱型图（Box Plot）：观察数据的分布范围和异常值
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Box Plot of Different Distributions")
plt.savefig(os.path.join(save_dir, "boxplot.png"), dpi=300, bbox_inches="tight")
plt.show()
```

```python
# 📊 3️⃣ 散点图矩阵（Pair Plot）：查看变量之间的关系
sns.pairplot(df, diag_kind="hist")  # 直方图作为对角线
plt.suptitle("Scatter Plot Matrix for Different Distributions", y=1.02)
plt.savefig(os.path.join(save_dir, "pairplot.png"), dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Create a 2x3 subplot layout
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # 2 rows, 3 columns layout

# Plot histograms for each distribution
columns = df.columns
for i, col in enumerate(columns):
    row, col_num = divmod(i, 3)  # Compute row and column index
    df[col].hist(bins=30, alpha=0.7, grid=False, edgecolor="black", ax=axes[row, col_num])
    axes[row, col_num].set_title(f"Histogram of {col}")

# Plot the Box Plot in the last position (bottom right)
sns.boxplot(data=df, ax=axes[1, 2])
axes[1, 2].set_title("Box Plot of Distributions")

# Remove the empty subplot to avoid blank space
fig.delaxes(axes[1, 1])  # Remove the empty subplot at (1,1)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Save the figure
plt.savefig(os.path.join(save_dir, "fixed_subplot_visualization.png"), dpi=300, bbox_inches="tight")

# Display the plot
plt.show()
```
